Contingency Table using CrossTab method of pandas

```python
import pandas as pd 
```

```python
df = pd.read_excel("./Country_data.xlsx");
df
```

```python
pd.crosstab(df.Nationality , df.Handedness )
```

```python
pd.crosstab(df.Sex, df.Handedness, margins=True)
```

```python
pd.crosstab([df.Sex, df.Nationality], df.Handedness, margins=True)
```

```python
pd.crosstab([df.Nationality, df.Name], [df.Sex, df.Handedness], margins=True)
```

```python
pd.crosstab(df.Nationality, df.Handedness, margins=True)
```

```python
pd.crosstab(df.Nationality, df.Handedness, margins=True, normalize='index')
```

```python
import numpy as np
```

```python
pd.crosstab(df.Sex, df.Handedness, values=df.Age, aggfunc=np.average)
```
