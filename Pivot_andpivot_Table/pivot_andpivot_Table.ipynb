PIVOT AND PIVOT TABLE 

```python
import pandas as pd 
df = pd.read_excel("./City_weather_data.xlsx")
df
```

```python
df.pivot(index = "Date", columns="City", values="Humidity")
```

```python
df.pivot(index = "Date", columns="City")
```

```python
df2 = pd.read_excel("./City_weather_data.xlsx", sheet_name="Sheet2");
df2
```

```python
df2.pivot_table(index="City", columns="Date", aggfunc="sum")
```

The many more aggregate function 

```python
df2.pivot_table(index="City", columns="Date", aggfunc="min")
```

```python
df2.pivot_table(index="City", columns="Date", margins=True)
```

GROUPER FUNCTION : to aggregate based on index frequency like here it is data 

```python
df3 = pd.read_excel("./City_weather_data.xlsx", sheet_name="Sheet3")
df3
```

```python
df3.pivot_table(index=pd.Grouper(freq="M", key="Date"), columns="City")
```
