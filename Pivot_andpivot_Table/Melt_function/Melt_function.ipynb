Reshape DataFrame using melt function

```python
import pandas as pd 
```

```python
df = pd.read_excel("./citywiseWeatherdata.xlsx");
df
```

```python
new_df = pd.melt(df, id_vars=["day"], var_name="city", value_name="Temperature")
new_df
```
