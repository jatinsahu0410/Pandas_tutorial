```python
import pandas as pd 
import numpy as np
```

```python
df = pd.read_excel("./weather2.xlsx");
df
```

```python
new_df = df.replace(-99999, np.NaN)
new_df
```

```python
new_df = df.replace([-99999, -88888], np.NaN)
new_df
```

```python
new_df = df.replace({
    'Temperature': [-99999, -88888],
    'windspeed' : -99999,
    'Event' : 'No event'
}, np.NaN);
new_df
```

```python
new_df = df.replace({
    -99999 : np.NaN,
    -88888 : np.NaN,
    'No event': 'Sunny'
})

new_df
```

we want to remove all the unit of measures 

```python
new_df = df.replace({
    'Temperature': '[A-Za-z]',
    'windspeed': '[A-Za-z]',
    }, '', regex=True);
new_df
```
